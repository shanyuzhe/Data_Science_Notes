### 理解 `re.search()`

`re.search(pattern, string, flags=0)` 会扫描 `string`，查找正则表达式 `pattern` 第一次匹配的位置。如果找到匹配项，它将返回一个 `match object`；否则，它将返回 `None`。

**与 `re.match()` 的主要区别：**

*   `re.match()` 只检查字符串**开头**的匹配项。
*   `re.search()` 会**扫描整个字符串**以查找第一个匹配项。

**参数：**

1.  **`pattern`**：这是您要查找的正则表达式。它可以是一个简单的字符串，也可以是一个更复杂的模式。
2.  **`string`**：这是您要在其中搜索的字符串。
3.  **`flags`**（可选）：修改模式解释方式的标志（例如，`re.IGNORECASE` 用于不区分大小写的匹配，`re.MULTILINE` 用于多行匹配）。

**返回值：**

*   如果找到匹配项，它将返回一个 **Match Object**。此对象包含有关匹配项的信息，例如匹配的字符串（`.group()`）、其开始和结束位置（`.start()`、`.end()`）等。
*   如果未找到匹配项，它将返回 `None`。


```python
import re

text = "The quick brown fox jumps over the lazy dog."

# 示例 1: 搜索一个简单的词
match1 = re.search("fox", text)
if match1:
    print(f"'fox' 找到: {match1.group()} 在位置 {match1.start()}-{match1.end()}")
else:
    print("'fox' 未找到")

# 示例 2: 搜索一个存在的模式
match2 = re.search("quick (.*?) fox", text)
if match2:
    print(f"模式 'quick (.*?) fox' 找到: {match2.group()}")
    print(f"捕获组: {match2.group(1)}") # (.*?) 非贪婪地捕获任何内容
else:
    print("模式 'quick (.*?) fox' 未找到")

# 示例 3: 搜索一个不存在的模式
match3 = re.search("cat", text)
if match3:
    print(f"'cat' 找到: {match3.group()}")
else:
    print("'cat' 未找到")

# 示例 4: 使用标志 (不区分大小写的搜索)
text_case = "Hello World"
match4 = re.search("world", text_case, re.IGNORECASE)
if match4:
    print(f"不区分大小写的 'world' 找到: {match4.group()}")
else:
    print("不区分大小写的 'world' 未找到")
```

在密码验证问题中，`re.search()` 被用于高效地检查密码字符串中是否存在至少一个小写字母、一个大写字母、一个数字和一个特殊字符（来自 `$#@`）。这比手动迭代字符串中的每种字符类型要简洁得多。

**OR**

```python
def check(x):
    cnt = 6 <= len(x) and len(x) <= 12
    for i in x:
        if i.isupper():
            cnt += 1
            break
    for i in x:
        if i.islower():
            cnt += 1
            break
    for i in x:
        if i.isnumeric():
            cnt += 1
            break
    for i in x:
        if i == "@" or i == "#" or i == "$":
            cnt += 1
            break
    return (
        cnt == 5
    )  # counting if total 5 all conditions are fulfilled then returns True


s = input().split(",")
lst = filter(
    check, s
)  # Filter function pick the words from s, those returns True by check() function
print(",".join(lst))
```

**OR**

```python
import re
s = input().split(',')
lst = []
for i in s:
  if ' ' in i: continue
  if len(i)<6 or len(i)>12:
    continue
  cnt = 0
  cnt += bool(re.search("[0-9]",i))
  cnt += bool(re.search("[a-z]",i))
  cnt += bool(re.search("[A-Z]",i))
  cnt += bool(re.search("[$#@]",i))
  if cnt == 4:
    lst.append(i)
print(','.join(lst))
```

### **Question:**

> **_You are required to write a program to sort the (name, age, score) tuples by ascending order where name is string, age and score are numbers. The tuples are input by console. The sort criteria is:_**

- **_1: Sort based on name_**
- **_2: Then sort based on age_**
- **_3: Then sort by score_**

> **_The priority is that name > age > score._**

> **_If the following tuples are given as input to the program:_**


Tom,19,80

John,20,90

Jony,17,91

Jony,17,93

Json,21,85


> **_Then, the output of the program should be:_**


[('John', '20', '90'), ('Jony', '17', '91'), ('Jony', '17', '93'), ('Json', '21', '85'), ('Tom', '19', '80')]

**Solutions:**

```python

lst = []
while True:
  s = input().split(',')
  if not s[0]:#if not s这样写最后一组输入是[''],不是空列表，不能退出
    break
  lst.append(tuple(s))

# 修正1: 将 sorted() 的结果赋值回 lst
# 修正2: 将年龄 (x[1]) 和分数 (x[2]) 转换为整数进行排序
lst = sorted(lst, key=lambda x: (x[0], int(x[1]), int(x[2])))
print(lst)
```
